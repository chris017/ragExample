# Inspect ChromaDB
Explore what's stored in the vector database.

```python
from pathlib import Path
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings

CHROMA_DIR = 'chroma_db'
EMBED_MODEL = 'all-MiniLM-L6-v2'

embeddings = HuggingFaceEmbeddings(model_name=EMBED_MODEL)
db = Chroma(persist_directory=CHROMA_DIR, embedding_function=embeddings, collection_name='rag_docs')
print(f'Total vectors stored: {db._collection.count()}')
```

## All chunks

```python
data = db._collection.get(include=['documents', 'metadatas'])

for i, (doc, meta) in enumerate(zip(data['documents'], data['metadatas'])):
    source = Path(meta.get('source', 'unknown')).name
    page = meta.get('page', '')
    label = f"{source} (p.{page + 1})" if page != '' else source
    print(f'--- Chunk {i+1} | {label} ---')
    print(doc)
    print()
```

## Test a similarity search

```python
query = 'What are the main skills?'  # change this

results = db.similarity_search_with_score(query, k=3)
for doc, score in results:
    source = Path(doc.metadata.get('source', 'unknown')).name
    print(f'Score: {score:.4f} | Source: {source}')
    print(doc.page_content)
    print()
```
